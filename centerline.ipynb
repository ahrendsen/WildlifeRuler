```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from ultralytics import YOLO
from wildlife_ruler.datasets import Fish, Newts, SeaHorses, Snakes
from wildlife_ruler.measure import get_centerline, initialize_sam3
```

```python
root = "results"
os.makedirs(root, exist_ok=True)

root_fish = '/data/wildlife_datasets/ruler/BarraRulerDataset'
root_seahorses = '/data/wildlife_datasets/ruler/seahorse'
root_snakes = '/data/wildlife_datasets/ruler/snakes_kent'
root_newts = "/data/wildlife_datasets/newts_kent"

newts_model = YOLO(os.path.join("weights", "newt-detector.pt"))
```

```python
model, processor = initialize_sam3()
```

```python
dataset1 = Fish(root_fish, "fish", model, processor)
dataset2 = Newts(root_newts, newts_model)
dataset3 = SeaHorses(root_seahorses, "animal", model, processor)
dataset4 = Snakes(root_snakes, "snake", model, processor)
datasets = [dataset1, dataset2, dataset3, dataset4]
```

```python
for dataset in datasets:
    for i in range(20):
        image = dataset[i]
        centerline = dataset.extract_centerline(i)

        if centerline is not None:
            plt.figure()
            plt.imshow(image)    
            plt.plot(centerline[:,1], centerline[:,0], color="yellow")
            plt.savefig(f"{root}/{dataset.name}_{i}.png", dpi=300)
            plt.close()
```
